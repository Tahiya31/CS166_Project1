## Step 2: Learn how is an image constructed

RGB (Red, Green, Blue) is the most common color model used in digital images and computer vision. It's a way to represent and manipulate colors in images. In this description, we'll explore how an RGB image is constructed, pixel by pixel.

**The Pixel as the Building Block**

An image is a collection of tiny dots called pixels. Each pixel is like a microscopic canvas that carries a specific color.

**The Role of Red, Green, and Blue**

In RGB, every pixel is composed of three primary colors: Red, Green, and Blue. These three colors are chosen because they closely mimic the way human vision works.

**Color as Intensity**

In RGB, color is represented as intensity values for each of the three primary colors. Each color channel (Red, Green, Blue) can have an intensity value ranging from 0 to 255, making 256 possible values per channel.

**Mixing Colors**

By varying the intensity of each color channel, we can create millions of unique colors. For example:
Full intensity Red (255, 0, 0) creates pure red.
Full intensity Green (0, 255, 0) creates pure green.
Full intensity Blue (0, 0, 255) creates pure blue.
Combining Red and Blue (255, 0, 255) creates magenta, and so on.

**The Role of Pixels in an Image**

When you have millions of these RGB pixels arranged in rows and columns, you get a complete image.
Each pixel contributes to the overall image by specifying its color.


**Resolution and Size**

The resolution of an image is determined by the number of pixels it contains. A higher resolution image has more pixels and can represent more details.
Image size is the physical dimensions of the image when printed or displayed. It depends on both the resolution and the physical size of each pixel.

### Opening and displaying the image

```python
from PIL import Image

# Open an image from a file
image = Image.open(image_path)      ## this function requires an image file path

# Display the image
display(image)
```

### Get dimensions of the image

```python
print('Image dimensions are:', image.size)
```

The first value is the width and second value is height.

### Resize the image



```python
# Define the new dimensions
width, height = 200, 400

# Resize the image
resized_image = image.resize((width, height))

# Display the image
display(resized_image)
```

### Convert image to grayscale

Here we select `L` as the mode to initiate a grayscale image. Read more about different modes [here](https://pillow.readthedocs.io/en/stable/handbook/concepts.html#modes).

```python
# Convert the image to grayscale
gray_image = image.convert('L')

# Display the grayscale image
display(gray_image)
```

### Crop an image


```python
# Define the coordinates of the cropping box (left, upper, right, lower)
box = (100, 100, 300, 300)

# Crop the image
cropped_image = image.crop(box)

# Display the cropped image
display(cropped_image)
```

### Here

**Image:** The Image object you want to crop.

**box:** A tuple specifying the cropping box as (left, upper, right, lower)

**left:** The x-coordinate of the left edge of the cropping box.

**upper:** The y-coordinate of the upper edge of the cropping box.

**right:** The x-coordinate of the right edge of the cropping box.

**lower:** The y-coordinate of the lower edge of the cropping box.
